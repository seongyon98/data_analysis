## 원본 데이터 "olist_orders_dataset" 데이터 전처리 


### 가설 : 해안 지역에 가까울수록 스포츠 관련 용품 구매율이 높다. 

```python
## olist_orders_dataset 데이터 불러오기
import pandas as pd

orders = pd.read_csv('Data/olist_orders_dataset.csv')
```

```python
## 기본정보 확인

orders.info()
```

```python
## 기본정보 확인


orders
```

```python
### 결측치 확인
orders.isnull().sum()
```

```python
### "order_status" 값들

order_status_names = orders["order_status"].unique()
order_status_names
```

```python
### "order_status"에서 필요한 "delivered" 값만 필터링 처리

orders_filtered = orders[orders["order_status"].isin(["delivered"])]

orders_filtered
```

```python
### 필터링한 데이터 결측치 확인
orders_filtered.isnull().sum()
```

```python
### "orders_filtered" 원데이터를 복사 ; 새로운 df "new_orders" 생성

new_orders = orders_filtered.copy()

new_orders
```

```python
### "new_orders" 결측치 가진 행 제거

new_orders.dropna(inplace=True)

new_orders

### "new_orders" 중복된 행 제거

new_orders.drop_duplicates(inplace=True)

new_orders
```

```python
### "new_orders" 결측치 다시 확인

new_orders.isnull().sum()
```

```python
### 날짜 형식 변환
new_orders['order_purchase_timestamp'] = pd.to_datetime(new_orders['order_purchase_timestamp'])
new_orders['order_approved_at'] = pd.to_datetime(new_orders['order_approved_at'])
new_orders['order_delivered_carrier_date'] = pd.to_datetime(new_orders['order_delivered_carrier_date'])
new_orders['order_delivered_customer_date'] = pd.to_datetime(new_orders['order_delivered_customer_date'])
new_orders['order_estimated_delivery_date'] = pd.to_datetime(new_orders['order_estimated_delivery_date'])

new_orders.info()
```

```python
def save_dataframe(dataframe: pd.DataFrame):
    dataframe.to_csv("new_order.csv")

save_dataframe(new_orders)

```
